```python
def enumerate_values(final_train, cols_to_convert):
    value_to_int = {}
    for col in cols_to_convert:
        unique_values = final_train[col].unique()
        for value in unique_values:
            if pd.notnull(value) and value not in value_to_int:
                value_to_int[value] = len(value_to_int) + 1
    value_to_int[np.nan] = 0  # Assigning 0 to NaN

    for col in cols_to_convert:
        final_train[col] = final_train[col].map(value_to_int)
 
    return final_train

final_train = enumerate_values(final_train, cols_to_convert)
```

```python
def enumerate_values(final_test, cols_to_convert):
    value_to_int = {}
    for col in cols_to_convert:
        unique_values = final_test[col].unique()
        for value in unique_values:
            if pd.notnull(value) and value not in value_to_int:
                value_to_int[value] = len(value_to_int) + 1
    value_to_int[np.nan] = 0  # Assigning 0 to NaN

    for col in cols_to_convert:
        final_test[col] = final_test[col].map(value_to_int)
 
    return final_test

final_test = enumerate_values(final_test, cols_to_convert)
```
